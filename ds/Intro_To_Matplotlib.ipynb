```python
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
_=plt.axis('Off')

plt.subplot(222)
plt.imshow(mpimg.imread('children.jpg'))
_=plt.axis('Off')

plt.subplot(223)
plt.imshow(mpimg.imread('workplace.jpg'))
_=plt.axis('Off')

plt.subplot(224)
plt.imshow(mpimg.imread('business.jpg'))
_=plt.axis('Off')
```

```python
fig = plt.figure(figsize = (8,6))

for idx, img in enumerate(['business.jpg', 'office.jpg', 'workplace.jpg', 'children.jpg']):
  plt.subplot(220 + idx + 1)
  plt.imshow(mpimg.imread(img))
  _=plt.axis('Off')
```
