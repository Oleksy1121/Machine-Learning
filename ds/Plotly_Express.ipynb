```python
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')
```

```python
df = fetch_financial_data()
df.head()
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
px.area(data_frame = df, x = 'Date', y='Close', log_y=True, title = 'Wykres spółki amazon')
```

Mapy

```python
df = pd.read_csv('https://ml-repository-krakers.s3-eu-west-1.amazonaws.com/plotly-course/us-cities-top-1k.csv')
df.head()
```

```python
fig = px.scatter_mapbox(df,lat='lat',lon='lon',hover_name='City',hover_data=['State','Population'])
fig.show()
```

```python
fig = px.scatter_mapbox(df,lat='lat',lon='lon',hover_name='City',hover_data=['State','Population'])
fig.update_layout(mapbox_style='carto-positron')
fig.show()
```

```python
fig = px.scatter_mapbox(df,lat='lat',lon='lon',hover_name='City',hover_data=['State','Population'],zoom=3)
fig.update_layout(mapbox_style='carto-positron', margin={'r':10,'t':10, 'l':10, 'b':10})
fig.show()
```

```python
fig = px.scatter_mapbox(df,lat='lat',lon='lon',hover_name='City',hover_data=['State','Population'],zoom=3, size='Population', color = 'State')
fig.update_layout(mapbox_style='carto-positron', margin={'r':10,'t':10, 'l':10, 'b':10})
fig.show()
```
