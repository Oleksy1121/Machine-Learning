Grupowanie danych

```python
# SELECT CustomerID, count(*) FROM retial GROUP BY CustomerID;
df.groupby('CustomerID').size()
```

```python
df.head(1)
```

```python
df['Revenue'] = df.Quantity * df.UnitPrice
df.head(1)
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM retial GROUP BY CustomerID;
df.groupby('CustomerID').aggregate({'Revenue':np.mean, 'CustomerID' : np.size})
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM retial GROUP BY CustomerID;
df.groupby('CustomerID').aggregate({'Revenue':np.mean, 'CustomerID' : np.size}).rename(columns={'Revenue': 'RevenueAvg', 'CustomerID':'Customer_Count'})
```

```python
df.head(1)
```

```python
df['InvoicaDay'] = df.InvoiceDate.dt.day
df.head(1)
```

```python
# SELECT InvoiceDateDay, sum(Revenue) FROM retail GROUP BY InvoiceDateDay;

results = df.groupby('InvoicaDay').aggregate({'Revenue':np.sum}).rename(columns={'Revenue':'RevenueSum', 'InvoicaDay':'DaysCount'})
```
