```python
import plotly
import plotly.graph_objects as go
```

```python
plotly.__version__
```

```python
fig = go.Figure(
    data=go.Bar(y=[1,4,2,5])
)

fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[1,4,2,5]),
    layout=go.Layout(title= {'text': 'Wykres słupkowy'})
)

fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[1,4,2,5]),
    layout=go.Layout(title_text = 'Wykres słupkowy')
)

fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[1,4,2,5]),
    layout=go.Layout(title_text = 'Wykres słupkowy', width = 800)
)

fig.write_html('wykres.html') #Tworzy plik html z wykresem
fig.show()
```

```python
from seaborn import load_dataset
import pandas as pd
df = load_dataset('diamonds')
df.head()
```

```python
df.info()
```

```python
dff = df.cut.value_counts()
dff
```

```python
dff = dff.reset_index()
```

```python
dff.columns = ['Type', 'Count']
dff.head()
```

```python
fig = go.Figure(
    data = go.Pie(labels = dff.Type, values = dff.Count),
    layout = go.Layout(title_text = 'Wykres kołowy', width = 500)
)

fig.show()
```

```python
fig = go.Figure(
    data = go.Pie(labels = dff.Type, values = dff.Count, hole =0.5),
    layout = go.Layout(title_text = 'Wykres kołowy', width = 500)
)

fig.show()
```

Diagram sankey'a

```python
data = [go.Sankey(node = dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019']),
                  link=dict(source = [0, 0, 1, 1], #indeks odpowiadający etykiecie
                            target = [2, 3, 2 ,3],
                            value =  [60,25,5,10]))] # wartość przepłytu

fig=go.Figure(data=data,
              layout = go.Layout(width=800))
fig.show()
```

```python
data = [go.Sankey(node = dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019', 'Nonchurn_2020', 'Churn_2020']),
                  link=dict(source = [0, 0, 1, 1, 2, 3], # Przepływ od
                            target = [2, 3, 2 ,3, 4, 5], # Przepływ do
                            value =  [60,25,5,10,65,35]))] # Wartość przepłytu

fig=go.Figure(data=data,
              layout = go.Layout(width=800))
fig.show()
```

```python
import pandas as pd
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
```

```python
df.head()
```

```python
df.reset_index()
```

```python
dff = df.query('Date > "2023-01-01"')
```

```python
dff.head()
```

```python
dff = dff.reset_index()
```

```python
dff.head()
```

```python
data = go.Candlestick(open = dff.Open, close = dff.Close, high = dff.High, low = dff.Low, x=dff.Date)

fig = go.Figure(data=data)
fig.show()
```
